```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
orders = pd.read_csv('./data/Orders.csv', index_col=0)
orders.columns
```

```python
returns = pd.read_csv('./data/Returns.csv', index_col=0)
returns.columns
```

```python
orders['Sales'] = orders['Sales'].str.replace('$', '')
```

```python
orders['Profit'] = orders['Profit'].str.replace('$', '')
```

```python
orders['Sales'] = orders['Sales'].str.replace(',', '')
```

```python
orders['Profit'] = orders['Profit'].str.replace(',', '')
```

```python
orders['Sales'] = pd.to_numeric(orders['Sales'])
```

```python
orders['Profit'] = pd.to_numeric(orders['Profit'])
```

```python
orders.groupby('Order.Date')[['Order.Date', 'Quantity']].agg({'Quantity':'sum'})
```

```python
orders['month'] = pd.DatetimeIndex(orders['Order.Date']).month
```

```python
plt.hist(orders.groupby('Order.Date')[['Order.Date', 'Quantity']].agg({'Quantity':'sum'}))
```

```python
orders.groupby('month')['Quantity'].agg({'Quantity':'sum'}).plot(kind = 'bar')
```

```python
#the amount of order will increase when the weather goes colder
```

```python
orders.groupby(['month','Category'])['Quantity'].agg({'Quantity':'sum'}).plot(kind = 'bar', figsize=(20,10))
```

```python
#all of the order incrase along with the season.
```
